```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Test different values of k
results = []
quantitative_df_copy = quantitative_df.copy()
quantitative_df_old=quantitative_df.copy()
for k in range(1, 21):  # Try k = 1 to 20
    # Copy the scaled data again for each loop
    test_data = quantitative_df_copy[quantitative_df_copy['MRI_Track_Age_at_Scan'].notnull()].copy()

    # Randomly mask 10% of values in 'MRI_Track_Age_at_Scan'
    np.random.seed(42)  # for reproducibility
    mask_indices = test_data.sample(frac=0.1).index
    true_values = test_data.loc[mask_indices, 'MRI_Track_Age_at_Scan'].copy()
    test_data.loc[mask_indices, 'MRI_Track_Age_at_Scan'] = np.nan

    # Apply KNN Imputer with current k
    knn_imputer = KNNImputer(n_neighbors=k)
    imputed_array = knn_imputer.fit_transform(test_data)

    # Extract imputed values
    row_positions = test_data.index.get_indexer(mask_indices)
    col_position = test_data.columns.get_loc('MRI_Track_Age_at_Scan')
    imputed_values = imputed_array[row_positions, col_position]

    # Calculate errors
    mae = mean_absolute_error(true_values, imputed_values)
    rmse = np.sqrt(mean_squared_error(true_values, imputed_values))

    results.append((k, mae, rmse))

# Convert results to DataFrame for easy analysis
results_df = pd.DataFrame(results, columns=['k', 'MAE', 'RMSE'])
print(results_df)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(results_df['k'], results_df['MAE'], marker='o', label='MAE')
plt.plot(results_df['k'], results_df['RMSE'], marker='x', label='RMSE')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Error')
plt.title('KNN Imputation Error vs k')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
from scipy.optimize import differential_evolution
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix
import pandas as pd

# Assuming X_train, y_train are already defined

# Custom evaluation metric
def custom_metric(estimator, X_val, y_val):
    # Predict
    y_pred = estimator.predict(X_val)
    
    # True Positives (TP) for "Sex_F" (assuming "Sex_F" is at column 0)
    sex_f_tp = np.sum((y_pred[:, 0] == 1) & (y_val.iloc[:, 0] == 1))
    
    # True Negatives (TN) for "ADHD_Outcome" (assuming "ADHD_Outcome" is at column 1)
    adhd_outcome_tn = np.sum((y_pred[:, 1] == 0) & (y_val.iloc[:, 1] == 0))

    return sex_f_tp, adhd_outcome_tn

# Model evaluation function for Differential Evolution
def evaluate_model(params):
    n_estimators, max_depth, max_features_index = params
    max_features_options = ['sqrt', 'log2']
    
    # Make sure max_features_index is either 0 or 1
    max_features_index = int(np.clip(round(max_features_index), 0, 1))
    max_features = max_features_options[max_features_index]

    # Create and fit the model
    model = MultiOutputClassifier(RandomForestClassifier(
        n_estimators=int(n_estimators),
        max_depth=int(max_depth),
        max_features=max_features,
        random_state=42,
        n_jobs=-1
    ))

    pipeline = Pipeline([
        ("clf", model)
    ])
    
    # Fit the model
    pipeline.fit(X_train, y_train)

    # Evaluate the model using the custom metric
    sex_f_tp, adhd_outcome_tn = custom_metric(pipeline, X_train, y_train)

    # Print metrics for each generation
    print(f"Generation - Params: {params}, TP (Sex_F): {sex_f_tp}, TN (ADHD_Outcome): {adhd_outcome_tn}")

    # Here we combine TP and TN to create a single "fitness" score
    fitness = sex_f_tp + adhd_outcome_tn
    return -fitness  # Return negative because differential_evolution minimizes the function

# Set bounds for the parameters
bounds = [
    (5, 100),     # n_estimators
    (1, 50),      # max_depth
    (0, 1.99),    # max_features_index (0 for "sqrt", 1 for "log2")
]

# Run differential evolution with the custom callback to print each generation
result = differential_evolution(
    evaluate_model,
    bounds,
    strategy='best1bin',
    maxiter=50,
    popsize=25,
    tol=0.01,
    mutation=(0.5, 1),
    recombination=0.7,
    seed=42,
    polish=True,
    disp=True
)

# Extract best parameters and print them
best_params = result.x
n_estimators = int(best_params[0])
max_depth = int(best_params[1])
max_features_index = int(round(best_params[2]))

# Ensure max_features_index is within range [0, 1]
max_features_index = np.clip(max_features_index, 0, 1)
max_features = ['sqrt', 'log2'][max_features_index]

print("🧬 Best Hyperparameters:")
print(f"n_estimators: {n_estimators}")
print(f"max_depth: {max_depth}")
print(f"max_features: {max_features}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, hamming_loss, classification_report
import pandas as pd
import numpy as np

# ----------------------------
# 1. Stratified Train/Test Split
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    full_df, solutions_df, test_size=0.2, random_state=42, stratify=solutions_df['Sex_F']
)

# ----------------------------
# 2. Define Best RF Model
# ----------------------------
best_rf = RandomForestClassifier(
    n_estimators=900,
    criterion='gini',
    max_depth=100,
    max_features='sqrt',
    class_weight='balanced',  # Handles class imbalance
    random_state=42,
    n_jobs=-1
)
multi_rf = MultiOutputClassifier(best_rf)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', multi_rf)
])

# ----------------------------
# 3. Train and Predict
# ----------------------------
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

adhd_true = y_test.iloc[:, 0].values
sex_true = y_test.iloc[:, 1].values
adhd_pred = y_pred[:, 0]
sex_pred = y_pred[:, 1]

# ----------------------------
# 4. Evaluation
# ----------------------------
adhd_acc = accuracy_score(adhd_true, adhd_pred)
sex_acc = accuracy_score(sex_true, sex_pred)
exact_match_accuracy = accuracy_score(y_test, y_pred)
hamming = hamming_loss(y_test, y_pred)

f1_adhd = f1_score(adhd_true, adhd_pred)
f1_sex = f1_score(sex_true, sex_pred)

def custom_f1_score(true_adhd, pred_adhd, true_sex):
    weights = np.ones_like(true_adhd, dtype=float)
    weights[(true_adhd == 1) & (true_sex == 1)] = 2.0
    return f1_score(true_adhd, pred_adhd, sample_weight=weights)

weighted_f1_adhd = custom_f1_score(adhd_true, adhd_pred, sex_true)
final_custom_score = (weighted_f1_adhd + f1_sex) / 2

# ----------------------------
# 5. Display Results
# ----------------------------
print("📊 Accuracy Scores:")
print(f"  - ADHD Diagnosis Accuracy: {adhd_acc:.4f}")
print(f"  - Sex Prediction Accuracy: {sex_acc:.4f}")
print(f"  - Exact Match Accuracy: {exact_match_accuracy:.4f}")
print(f"  - Hamming Loss: {hamming:.4f}")

print("\n📈 F1 Scores:")
print(f"  - Standard F1 (ADHD): {f1_adhd:.4f}")
print(f"  - Standard F1 (Sex): {f1_sex:.4f}")

print("\n🎯 Custom Leaderboard Score:")
print(f"  - Weighted F1 (ADHD, Female Emphasis): {weighted_f1_adhd:.4f}")
print(f"  - Weighted F1 (Sex): {f1_sex:.4f}")
print(f"  - Final Custom Score: {final_custom_score:.4f}")

print("\n📋 Classification Report for Sex:")
print(classification_report(sex_true, sex_pred, target_names=["Male", "Female"])) 
```
