```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Test different values of k
results = []
quantitative_df_copy = quantitative_df.copy()
quantitative_df_old=quantitative_df.copy()
for k in range(1, 21):  # Try k = 1 to 20
    # Copy the scaled data again for each loop
    test_data = quantitative_df_copy[quantitative_df_copy['MRI_Track_Age_at_Scan'].notnull()].copy()

    # Randomly mask 10% of values in 'MRI_Track_Age_at_Scan'
    np.random.seed(42)  # for reproducibility
    mask_indices = test_data.sample(frac=0.1).index
    true_values = test_data.loc[mask_indices, 'MRI_Track_Age_at_Scan'].copy()
    test_data.loc[mask_indices, 'MRI_Track_Age_at_Scan'] = np.nan

    # Apply KNN Imputer with current k
    knn_imputer = KNNImputer(n_neighbors=k)
    imputed_array = knn_imputer.fit_transform(test_data)

    # Extract imputed values
    row_positions = test_data.index.get_indexer(mask_indices)
    col_position = test_data.columns.get_loc('MRI_Track_Age_at_Scan')
    imputed_values = imputed_array[row_positions, col_position]

    # Calculate errors
    mae = mean_absolute_error(true_values, imputed_values)
    rmse = np.sqrt(mean_squared_error(true_values, imputed_values))

    results.append((k, mae, rmse))

# Convert results to DataFrame for easy analysis
results_df = pd.DataFrame(results, columns=['k', 'MAE', 'RMSE'])
print(results_df)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(results_df['k'], results_df['MAE'], marker='o', label='MAE')
plt.plot(results_df['k'], results_df['RMSE'], marker='x', label='RMSE')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Error')
plt.title('KNN Imputation Error vs k')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

def impute_quantitative_data(df, target_column='MRI_Track_Age_at_Scan', knn_neighbors=9):
    """
    Applies KNN imputation to the specified column and fills the remaining missing values with median.
    
    Parameters:
        df (pd.DataFrame): The quantitative DataFrame with participant_id as the index.
        target_column (str): The column to impute using KNN.
        knn_neighbors (int): Number of neighbors for KNN imputation.
    
    Returns:
        pd.DataFrame: The imputed DataFrame.
        pd.Series: Missing value count after imputation.
    """
    df_copy = df.copy()
    scaled_df = df.copy()
    
    # Scale all columns except the target
    columns_to_scale = scaled_df.drop(columns=[target_column]).columns
    scaler = StandardScaler()
    scaled_df[columns_to_scale] = scaler.fit_transform(scaled_df[columns_to_scale])
    
    # Apply KNN imputer to the target column
    knn_imputer = KNNImputer(n_neighbors=knn_neighbors)
    scaled_df[target_column] = knn_imputer.fit_transform(scaled_df[[target_column]])
    
    # Replace original target column with imputed values
    df[target_column] = scaled_df[target_column]
    
    # Fill remaining NaNs with column medians
    df.fillna(df.median(), inplace=True)
    
    # Return both the cleaned DataFrame and the missing values summary
    missing_summary = df.isnull().sum()
    print(f"Missing values after KNN and median imputation:\n{missing_summary}")
    
    return df, missing_summary
quantitative_df_imputed, missing_summary = impute_quantitative_data(quantitative_df)
```

```python
import pandas as pd

def cap_outliers_iqr(df):
    """
    Detects outliers using the IQR method and applies capping and flooring.
    
    Parameters:
        df (pd.DataFrame): The DataFrame containing numerical data.
    
    Returns:
        pd.DataFrame: The capped DataFrame.
        pd.DataFrame: A summary of outlier count and percentage.
    """
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = (df < lower_bound) | (df > upper_bound)
    outlier_counts = outliers.sum()
    outlier_percentage = (outlier_counts / len(df)) * 100

    outlier_summary = pd.DataFrame({
        'Outlier Count': outlier_counts,
        'Outlier Percentage (%)': outlier_percentage
    })

    # Apply capping and flooring
    df_capped = df.copy()
    for col in df.columns:
        df_capped[col] = df_capped[col].clip(lower=lower_bound[col], upper=upper_bound[col])

    print("Outlier Summary (Count and Percentage):")
    print(outlier_summary)

    return df_capped, outlier_summary
quantitative_df_capped, outlier_summary = cap_outliers_iqr(quantitative_df)
```

```python
import pandas as pd

def impute_categorical(categorical_df, columns_to_impute_with_minus_1):
    """
    Imputes missing values in a categorical DataFrame.
    - Selected columns are imputed with -1.
    - The rest are imputed with the mode (most frequent value).
    
    Parameters:
        categorical_df (pd.DataFrame): The DataFrame containing categorical features.
        columns_to_impute_with_minus_1 (list): Columns to fill with -1.
    
    Returns:
        pd.DataFrame: DataFrame with missing values handled.
    """
    df = categorical_df.copy()

    # Impute selected columns with -1
    df[columns_to_impute_with_minus_1] = df[columns_to_impute_with_minus_1].fillna(-1)

    # Impute remaining missing values with mode
    columns_with_missing = df.columns[df.isnull().any()]
    for column in columns_with_missing:
        if column not in columns_to_impute_with_minus_1:
            mode_value = df[column].mode()[0]
            df[column] = df[column].fillna(mode_value)

    return df
columns_to_impute_with_minus_1 = ['Barratt_Barratt_P2_Edu', 'Barratt_Barratt_P2_Occ']
categorical_df = impute_categorical(categorical_df, columns_to_impute_with_minus_1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def analyze_and_group_categorical(categorical_df, threshold=1, plot_columns=9):
    """
    Analyzes and transforms categorical columns:
    - Prints unique counts.
    - Prints the lowest value percentage for each column.
    - Plots value counts for the first N columns.
    - Groups rare categories (below threshold %) into 'Other'.
    
    Parameters:
        categorical_df (pd.DataFrame): The DataFrame with categorical data.
        threshold (float): Percentage below which categories are grouped into 'Other'.
        plot_columns (int): Number of columns to plot.
    
    Returns:
        pd.DataFrame: Transformed DataFrame with rare categories grouped.
    """
    df = categorical_df.copy()

    # Unique value counts
    unique_counts = df.nunique()
    print("Unique values per column:")
    print(unique_counts)

    # Lowest value percentage per column
    lowest_percentage = {}
    for col in df.columns:
        value_counts = df[col].value_counts(normalize=True) * 100
        lowest_percentage[col] = value_counts.min()
    print("\nLowest percentage of a category in each column:")
    print(lowest_percentage)

    # Plotting value counts for the first few columns
    n_cols = 3
    n_rows = (plot_columns + n_cols - 1) // n_cols
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
    axes = axes.flatten()

    for i, col in enumerate(df.columns[:plot_columns]):
        df[col].value_counts().plot(kind='bar', color='skyblue', ax=axes[i])
        axes[i].set_title(f'Value Counts for {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Count')
        axes[i].tick_params(axis='x', rotation=45)

    # Hide any unused subplots
    for j in range(plot_columns, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

    # Group rare categories
    for col in df.columns:
        value_counts = df[col].value_counts(normalize=True) * 100
        rare_categories = value_counts[value_counts < threshold].index
        df[col] = df[col].replace(rare_categories, 'Other')

    return df

categorical_df = analyze_and_group_categorical(categorical_df, threshold=1, plot_columns=9)
```

```python
import pandas as pd

def one_hot_encode_dataframe(df):
    """
    Converts all values in a DataFrame to string type and performs one-hot encoding.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing categorical features.
    
    Returns:
        pd.DataFrame: One-hot encoded DataFrame.
    """
    # Convert all entries to string type
    df_str = df.applymap(str)
    
    # Apply one-hot encoding
    encoded_df = pd.get_dummies(df_str)
    
    return encoded_df

encoded_df = one_hot_encode_dataframe(categorical_df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTENC

# Assume selected_df (features) and solutions_df (targets) are already defined

# 1. Split into X and y
X = selected_df.copy()
y = solutions_df.copy()

# 2. Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y["Sex_F"]
)

# 3. Define integer and binary features
integer_features = [
    'ColorVision_CV_Score', 'APQ_P_APQ_P_CP', 'APQ_P_APQ_P_ID', 'APQ_P_APQ_P_INV',
    'APQ_P_APQ_P_OPD', 'APQ_P_APQ_P_PM', 'APQ_P_APQ_P_PP', 'SDQ_SDQ_Conduct_Problems',
    'SDQ_SDQ_Difficulties_Total', 'SDQ_SDQ_Emotional_Problems', 'SDQ_SDQ_Externalizing',
    'SDQ_SDQ_Generating_Impact', 'SDQ_SDQ_Hyperactivity', 'SDQ_SDQ_Internalizing',
    'SDQ_SDQ_Peer_Problems', 'SDQ_SDQ_Prosocial'
]

binary_features = [
    'PreInt_Demos_Fam_Child_Ethnicity'
]

# 4. Scale only the integer features

features_to_scale = [col for col in X_train.columns if col not in binary_features]
scaler = StandardScaler()

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# Scale only non-binary features
X_train_scaled[features_to_scale] = scaler.fit_transform(X_train_scaled[features_to_scale])
X_test_scaled[features_to_scale] = scaler.transform(X_test_scaled[features_to_scale])

# 5. Prepare for SMOTENC
y_train_combined = y_train.astype(str).agg('-'.join, axis=1)

categorical_feature_indices = [i for i, col in enumerate(X_train_scaled.columns) if col in binary_features]

# 6. Apply SMOTENC
smote_nc = SMOTENC(categorical_features=categorical_feature_indices, random_state=42)
X_train_resampled, y_train_resampled_combined = smote_nc.fit_resample(X_train_scaled, y_train_combined)

# 7. Separate back y into two columns
y_train_resampled = y_train_resampled_combined.str.split('-', expand=True).astype(int)
y_train_resampled.columns = ['ADHD_Outcome', 'Sex_F']

# 8. Fix the integer columns (round because SMOTE introduces floats)
X_train_resampled_df = pd.DataFrame(X_train_resampled, columns=X_train_scaled.columns)
X_train_resampled_df[integer_features] = X_train_resampled_df[integer_features].round().astype(int)

# 9. One-hot encode binary features
X_train_final = pd.get_dummies(X_train_resampled_df, columns=binary_features, drop_first=False)
X_test_final = pd.get_dummies(X_test_scaled, columns=binary_features, drop_first=False)

# 10. Align train and test columns
X_test_final = X_test_final.reindex(columns=X_train_final.columns)

# 11. Assign final data
X_train = X_train_final
X_test = X_test_final
y_train = y_train_resampled

# 12. Print results
print("\nBalanced label counts:\n", pd.DataFrame(y_train).value_counts())
assert X_train_final.columns.equals(X_test_final.columns), "Columns do not match!"
```

```python
from sklearn.metrics import (
    make_scorer, f1_score, confusion_matrix, classification_report
)
from sklearn.model_selection import GridSearchCV, train_test_split

def custom_competition_metric(y_true, y_pred):
    # Ensure data is in int type for correct comparison
    adhd_true, sex_true = y_true[:, 0].astype(int), y_true[:, 1].astype(int)
    adhd_pred, sex_pred = y_pred[:, 0].astype(int), y_pred[:, 1].astype(int)

    # Initialize weights, giving double weight to Female ADHD cases (ADHD_Outcome=1, Sex_F=1)
    weights = np.ones_like(adhd_true, dtype=float)
    weights[(adhd_true == 1) & (sex_true == 1)] = 2.0  # Double weight for female ADHD

    # Calculate the true positives, false positives, and false negatives for ADHD
    tp_adhd = np.sum(((adhd_pred == 1) & (adhd_true == 1)) * weights)
    fp_adhd = np.sum(((adhd_pred == 1) & (adhd_true == 0)) * weights)
    fn_adhd = np.sum(((adhd_pred == 0) & (adhd_true == 1)) * weights)

    # Precision and Recall for ADHD
    precision_adhd = tp_adhd / (tp_adhd + fp_adhd + 1e-8)
    recall_adhd = tp_adhd / (tp_adhd + fn_adhd + 1e-8)
    f1_adhd = 2 * precision_adhd * recall_adhd / (precision_adhd + recall_adhd + 1e-8)

    # F1 score for Sex classification
    f1_sex = f1_score(sex_true, sex_pred, pos_label=1)

    # Return the average F1 score of both targets
    return (f1_adhd + f1_sex) / 2


custom_scorer = make_scorer(custom_competition_metric, greater_is_better=True)
# === 1. DATA PREPARATION ===
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2
from sklearn.metrics import f1_score, classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# === 1. DEFINE KERAS MODEL ===
def create_model():
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=l2(0.01)))
    model.add(Dense(32, activation='relu', kernel_regularizer=l2(0.01)))
    model.add(Dense(16, activation='relu', kernel_regularizer=l2(0.001)))
    model.add(Dense(2, activation='sigmoid'))  # Output layer for multi-output classification (ADHD and Sex)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    return model

# === 2. TRAIN THE MODEL ===
model = create_model()

# Fit the model on the training data
model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=1)

# === 3. PREDICTIONS ===
y_pred = model.predict(X_test)

# Assuming y_test is a DataFrame
y_test_np = y_test.values  # Convert y_test DataFrame to a NumPy array

f1_scores = []
for i, label in enumerate(["ADHD", "Sex (F)"]):
    # F1 score
    f1 = f1_score(y_test_np[:, i], y_pred[:, i] > 0.5)  # Apply threshold of 0.5 for binary classification
    f1_scores.append(f1)
    print(f"🔶 F1 Score for {label}: {f1:.4f}")

    # Accuracy
    accuracy = np.mean(y_test_np[:, i] == (y_pred[:, i] > 0.5))  # Apply threshold of 0.5 for binary classification
    print(f"🔷 Accuracy for {label}: {accuracy:.4f}")

# === 5. JOINT ACCURACY ===
joint_accuracy = np.mean(np.all(y_test_np == (y_pred > 0.5), axis=1))  # Apply threshold for joint accuracy
print(f"🔗 Joint Accuracy: {joint_accuracy:.4f}")
# === 6. REPORTS & CONFUSION MATRICES ===
for i, label in enumerate(["ADHD", "Sex (F)"]):
    print(f"\n📊 Classification Report for {label}:")
    print(classification_report(y_test_np[:, i], y_pred[:, i] > 0.5))  # Apply threshold for binary classification

    cm = confusion_matrix(y_test_np[:, i], y_pred[:, i] > 0.5)  # Apply threshold for confusion matrix
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"Confusion Matrix: {label}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.xticks([0.5, 1.5], ["0", "1"])
    plt.yticks([0.5, 1.5], ["0", "1"])
    plt.tight_layout()
    plt.show()

# === 7. SAVE THE MODEL ===
model.save('keras_model.h5')  # Saving the model in .h5 format

```
